```r
#load the psych package
library(tidyverse)
library(psych)


#load built in dataset 'sat.act'
dat = sat.act

#run describe on the dataset & save obj
datDscrb = data.frame(describe(dat))
```

```r
#Now run the code and include BOTH the code and output
library(knitr)
library(kableExtra)
library(IRdisplay)

#one liner
#datDscrb1 = round(datDscrb[, c("mean", "sd")], 3)


datDscrb %>%
    select("mean", "sd") %>%
    round(3) %>%
    kable(format = "html", escape = F, row.names = T, 
          col.names = c("Mean", "SD")) %>%
        kable_styling(bootstrap_options = c("hover", "condensed"), full_width = F) %>%
    as.character() %>%
    IRdisplay::display_html()
```
